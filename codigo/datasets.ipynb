```python
# Ordenar por fecha
df_startup_mmm['Calendar_Week'] = pd.to_datetime(df_startup_mmm['Calendar_Week'])
df_startup_mmm = df_startup_mmm.sort_values(by='Calendar_Week')
```

```python
#  Revisar las fechas y tipos de datos

# Sumamos todos los registros por dia
df_startup_mmm.drop('Division',axis=1,inplace=True)
df_startup_mmm = df_startup_mmm.groupby('Calendar_Week').sum()
df_startup_mmm.plot()
```

```python
# Suponiendo que df_startup_mmm es tu DataFrame y que el índice ya es de tipo DateTime
# Crear la variable binaria para noviembre y diciembre
df_startup_mmm['is_peak_season'] = df_startup_mmm.index.month.isin([11, 12])

# Convertir booleanos a enteros para la variable binaria
df_startup_mmm['is_peak_season'] = df_startup_mmm['is_peak_season'].astype(int)

# Ver algunos datos para confirmar
print(df_startup_mmm[['is_peak_season']])
```

```python
import pandas as pd
from statsmodels.tsa.stattools import ccf
import matplotlib.pyplot as plt

# Calcula la correlación cruzada entre 'Paid_Views' y 'Sales'
correlation = ccf(df_startup_mmm['Paid_Views'], df_startup_mmm['Sales'], adjusted=False)

# Graficar la correlación cruzada
plt.figure(figsize=(10, 5))
plt.stem(range(len(correlation)), correlation, use_line_collection=True)
plt.title('Correlación Cruzada entre Paid Views y Sales')
plt.xlabel('Lag')
plt.ylabel('Correlación Cruzada')
plt.show()
```

```python
# La gráfica que has generado muestra la correlación cruzada entre las vistas pagadas (Paid Views) y las ventas (Sales). Algunas observaciones son:

# Peaks destacados: Hay picos significativos en la correlación a varios lags, lo que podría sugerir que las vistas pagadas influyen en las ventas con un retraso específico. Los picos en lags bajos indican que el efecto es casi inmediato, mientras que los picos en lags más altos sugieren efectos retardados.

# Correlación negativa: A medida que el lag aumenta, parece que la correlación se vuelve negativa, aunque muy cercana a cero, lo cual podría no ser estadísticamente significativo sin una prueba de significancia adicional.

# Variabilidad: La variabilidad en la correlación a diferentes lags sugiere que la relación entre Paid Views y Sales puede ser más compleja y estar influenciada por otros factores o por la naturaleza dinámica de cómo se procesan y se ven afectadas las ventas.
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import ccf
import pandas as pd

# Suponiendo que 'df' es tu DataFrame

# Lista de columnas para comparar
columns = df_startup_mmm.columns.tolist()

# Crear una figura para contener todos los subplots
# Ajusta el tamaño aquí, por ejemplo (40, 40) o según sea necesario para tu pantalla
fig, axes = plt.subplots(nrows=len(columns), ncols=len(columns), figsize=(40, 40))

# Iterar sobre cada combinación de columnas
for i, col1 in enumerate(columns):
    for j, col2 in enumerate(columns):
        if i != j:
            # Calcular la correlación cruzada
            # Limita el número de lags a mostrar, por ejemplo, los primeros 50 lags
            corr = ccf(df_startup_mmm[col1], df_startup_mmm[col2], adjusted=False)[:50]
            # Graficar la correlación cruzada
            axes[i, j].stem(range(len(corr)), corr, use_line_collection=True)
            axes[i, j].set_title(f'CCF: {col1} vs {col2}')
            axes[i, j].set_ylim([-0.5, 0.5])  # Ajusta esto según tus datos
        else:
            # Eliminar el eje cuando no es necesario
            fig.delaxes(axes[i][j])

# Ajusta el espaciado para evitar la superposición de elementos
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, hspace=0.4, wspace=0.4)

plt.show()
```
